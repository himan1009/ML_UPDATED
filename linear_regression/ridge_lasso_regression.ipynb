## ðŸ”¥ Algerian Forest Fires Dataset

### ðŸ“„ Data Set Information:

The dataset includes **244 instances** that regroup data from two regions of Algeria:  
- **Bejaia** region (northeast Algeria)  
- **Sidi Bel-abbes** region (northwest Algeria)  

There are **122 instances** for each region.

The data covers the period from **June 2012 to September 2012**.  
It includes **11 attributes** and **1 output attribute** (the fire occurrence class).  

Out of the 244 instances:  
- **138 are classified as fire**  
- **106 are classified as not fire**

---

### ðŸ“Š Attribute Information:

1. **Date**: (DD/MM/YYYY)  
   - Day, month (`June` to `September`), year (2012)  
   - Weather data observations  

2. **Temp**: Temperature at noon (max temp) in Â°C â€” **range: 22 to 42**

3. **RH**: Relative Humidity in % â€” **range: 21 to 90**

4. **Ws**: Wind Speed in km/h â€” **range: 6 to 29**

5. **Rain**: Total daily rainfall in mm â€” **range: 0 to 16.8**

---

#### ðŸŒ² Fire Weather Index (FWI) Components:

6. **FFMC** (Fine Fuel Moisture Code) â€” **range: 28.6 to 92.5**

7. **DMC** (Duff Moisture Code) â€” **range: 1.1 to 65.9**

8. **DC** (Drought Code) â€” **range: 7 to 220.4**

9. **ISI** (Initial Spread Index) â€” **range: 0 to 18.5**

10. **BUI** (Buildup Index) â€” **range: 1.1 to 68**

11. **FWI** (Fire Weather Index) â€” **range: 0 to 31.1**

---

12. **Classes**:

### ðŸ§¾ Region-wise Dataset Split

The dataset contains a total of **244 rows**, with **122 instances for each region**:

- The **first 122 rows (index 0â€“121)** belong to the **Bejaia Region**
- The **next 122 rows (index 122â€“243)** belong to the **Sidi-Bel Abbes Region**

To clearly distinguish the source region of each record, we created a new column named **`Region`** with the following mapping:

- `1` â†’ **"Bejaia Region Dataset"**
- `2` â†’ **"Sidi-Bel Abbes Region Dataset"**

This column will help in region-wise analysis and visualization of fire occurrences.
